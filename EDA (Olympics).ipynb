```python
gold.Region.value_counts().reset_index(name="Medal").head()
```

```python
totalgold = gold.Region.value_counts().reset_index(name="Medal").head(5)
g = sns.catplot(x="index", y="Medal", data = totalgold, height = 5, palette = "rocket", kind = "bar" )
g.despine(left = True)
g.set_xlabels("Top 5 countries")
g.set_ylabels("Number of medals")
plt.title = ("gold medals per country")

```
